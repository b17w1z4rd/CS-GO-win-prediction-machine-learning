```python
import requests
url = "https://www.openml.org/data/download/22102255/dataset"
r= requests.get(url, allow_redirects=True)
```

```python
with open("dataset.txt","wb") as f:
    f.write(r.content)
```

```python
import pandas as pd
```

```python
data = []
with open("dataset.txt","r") as f:
    for line in f.read().split("\n"):
      if line.startswith("@") or line.startswith("%") or line == "":
        continue
      data.append(line)
```

```python
columns = []
with open("dataset.txt","r") as f:
    for line in f.read().split("\n"):
      if line.startswith("@ATTRIBUTE"):
        columns.append(line.split(" ")[1])
```

```python
data
```

```python
with open("df.csv","w") as f:
  f.write(",".join(columns)+"\n")
  f.write("\n")
  f.write("\n".join(data))
```

```python
df = pd.read_csv("df.csv")
df.columns = columns
df
```

```python
df['t_win'] = df.round_winner.astype("category").cat.codes
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns for correlation calculation
numeric_columns = df[columns + ["t_win"]].select_dtypes(include='number').columns
correlations = df[numeric_columns].corr()

print(correlations['t_win'].apply(abs).sort_values(ascending=False).iloc[:25])
```

```python
selected_columns=[]
for col in columns+["t_win"]:
    try:
      if abs(correlations['t_win'][col])>0.15:

        selected_columns.append(col)
    except:
        pass
df_selected = df[selected_columns]
```

```python
plt.figure(figsize=(18, 18))
sns.heatmap(df_selected.corr().sort_values(by="t_win"), annot=True, cmap='coolwarm')
```
